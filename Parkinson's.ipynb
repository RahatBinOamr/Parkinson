## import necessary libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load dataset 

```python
data = pd.read_csv('parkinsons.csv')
data.head(10)
```

### Check null data on this data set


```python
data.isnull()
```

### Remove null data on this dataset

```python
data = data.dropna()
data
```

### Question 1: What is the distribution of Parkinson's disease status in the dataset?

```python
status_counts = data['status'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(
    status_counts,
    labels=['Healthy', 'Parkinson\'s'],
    autopct='%1.1f%%',
    startangle=140,
    colors=['green', 'red'],
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Distribution of Parkinson's Disease Status", fontsize=14)
plt.show()
```

***Explanation:***
* Parkinson's patients: Majority of individuals in the dataset.
* Healthy individuals: Minority class.
* This skewed distribution indicates a class imbalance, which should be considered when building predictive models.

### Question 2: Which features are most correlated with Parkinson's disease status?

```python
numeric_data = data.select_dtypes(include=['number'])
correlation_with_status = numeric_data.corr()['status'].sort_values(ascending=False)
top_correlations = correlation_with_status.head(10)
plt.figure(figsize=(10, 6))
top_correlations.drop('status').plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top Features Correlated with Parkinson\'s Status', fontsize=14)
plt.ylabel('Correlation Coefficient', fontsize=12)
plt.xlabel('Features', fontsize=12)
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

***Explanation:***
* The most positively correlated feature with Parkinson’s disease is spread1 (0.56), followed by PPE (0.53).
* Features like spread2 and DFA also show moderate correlation.
* These features are important for diagnosing Parkinson's.

### Question 3: How do jitter and shimmer values differ between healthy and Parkinson's patients?

```python
healthy = data[data['status'] == 0]
parkinsons = data[data['status'] == 1]
jitter_cols = ['MDVP:Jitter(%)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP']
mean_jitter_healthy = healthy[jitter_cols].mean()
mean_jitter_parkinsons = parkinsons[jitter_cols].mean()
plt.figure(figsize=(10, 6))
x = range(len(jitter_cols))
width = 0.35
plt.bar([i - width / 2 for i in x], mean_jitter_healthy, width, label='Healthy', color='green', alpha=0.7)
plt.bar([i + width / 2 for i in x], mean_jitter_parkinsons, width, label='Parkinson\'s', color='red', alpha=0.7)
plt.xticks(x, jitter_cols, rotation=45, ha='right')
plt.title('Comparison of Jitter Metrics', fontsize=14)
plt.ylabel('Mean Value', fontsize=12)
plt.xlabel('Jitter Features', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

***Explanation:***
* Parkinson’s patients exhibit higher jitter values, indicating instability in pitch.
* This aligns with known symptoms of the disease.

### Question 4: How do shimmer values differ between healthy and Parkinson's patients?

```python
shimmer_cols = ['MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:DDA']
mean_shimmer_healthy = healthy[shimmer_cols].mean()
mean_shimmer_parkinsons = parkinsons[shimmer_cols].mean()
plt.figure(figsize=(10, 6))
x = range(len(shimmer_cols))
width = 0.35
plt.bar([i - width / 2 for i in x], mean_shimmer_healthy, width, label='Healthy', color='green', alpha=0.7)
plt.bar([i + width / 2 for i in x], mean_shimmer_parkinsons, width, label='Parkinson\'s', color='red', alpha=0.7)
plt.xticks(x, shimmer_cols, rotation=45, ha='right')
plt.title('Comparison of Shimmer Metrics', fontsize=14)
plt.ylabel('Mean Value', fontsize=12)
plt.xlabel('Shimmer Features', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

***Explanation:***
* Parkinson’s patients have higher shimmer values, indicating amplitude irregularities.
* This is another significant vocal indicator for identifying the disease.


### Question 5: How does the Harmonics-to-Noise Ratio (HNR) vary between healthy and Parkinson's groups?

```python
plt.figure(figsize=(8, 5))
plt.hist(healthy['HNR'], bins=30, alpha=0.7, label='Healthy', color='green')
plt.hist(parkinsons['HNR'], bins=30, alpha=0.7, label='Parkinson\'s', color='red')
plt.title('Harmonics-to-Noise Ratio (HNR) Distribution', fontsize=14)
plt.xlabel('HNR', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

***Explanation:***
* Healthy individuals have a higher mean HNR, reflecting clearer voice quality.
* Parkinson’s patients exhibit lower HNR values, indicative of voice disturbances.
